## 1.4. Clean and audit the data in restaurants_data

```python
# build a dataframe with some attributes

selectedColumns = ['id', 'name', 'url','rating','price', 'location', 'display_phone', 'review_count']
restaurantsDataFrame = pd.DataFrame(restaurants_data_1['businesses'], columns = selectedColumns)
restaurantsDataFrame
```

```python
# Maximum number of results to return by yelp API is 50. So, try more times to get more data

for i in range(1,19):
    PARAMETERS = {'term': 'restaurants',
              'limit': 50,
              'offset': 50*i,
              'location':'Boston'}
    response = requests.get(url = ENDPOINT,
                        params = PARAMETERS,
                        headers = HEADERS)
    restaurants_data = response.json()
    newDataFrame = pd.DataFrame(restaurants_data['businesses'], columns = selectedColumns)
    restaurantsDataFrame = restaurantsDataFrame.append(newDataFrame,ignore_index = True)

print(restaurantsDataFrame)
```

# 3. DATASOURCE 3 - Using Web Scraping with Beautifulsoup

```python
# import liabraries for web scraping
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
# Fetch the tags from the website
r = requests.get("https://www.yelp.com/search?cflt=restaurants&find_loc=Boston%2C+MA")
c = r.content
c
```

```python
# Display the above format to more readable format using html parser
soup = BeautifulSoup(c,"html.parser")
print(soup.prettify)
```

```python
# Scrap the name and category of restaurants in boston on yelp
testname = []
testcategory = []
attributes = soup.find_all("div",{"class":"lemon--div__373c0__1mboc mainAttributes__373c0__1r0QA arrange-unit__373c0__o3tjT arrange-unit-fill__373c0__3Sfw1 border-color--default__373c0__3-ifU"})
for i in attributes:
    temp_name = i.find("a").text
    testname.append(temp_name)
    temp_category = i.find("span",{"class":"lemon--span__373c0__3997G text__373c0__2Kxyz text-color--black-extra-light__373c0__2OyzO text-align--left__373c0__2XGa-"}).text
    testcategory.append(temp_category)
print(testname)
print(testcategory)
```

```python
# clean testcategory
cleancategory = []
for i in testcategory:
    if (',' in i):
        cleancategory.append(i[:-2])
    else:
        cleancategory.append(i)

print(cleancategory)
```

```python
# build a dataframe
df_category = pd.DataFrame(testname, columns=['testname'])
df_category = pd.concat([df_category, pd.DataFrame(cleancategory,columns=['testcategory'])],axis=1)

print(df_category)
```

```python
# try a loop to scrape data from more pages
name = []
category = []

for i in range(0,33): 
    url = "https://www.yelp.com/search?cflt=restaurants&find_loc=Boston%2C%20MA&start=" + str(i*30)
    req = requests.get(url)
    con = req.content
    Beautisoup = BeautifulSoup(con,"html.parser")
    attris = Beautisoup.find_all("div",{"class":"lemon--div__373c0__1mboc mainAttributes__373c0__1r0QA arrange-unit__373c0__o3tjT arrange-unit-fill__373c0__3Sfw1 border-color--default__373c0__3-ifU"})
    for attri in attris:
        temp_name = attri.find("a").text
        name.append(temp_name)
        temp_category = attri.find("span",{"class":"lemon--span__373c0__3997G text__373c0__2Kxyz text-color--black-extra-light__373c0__2OyzO text-align--left__373c0__2XGa-"}).text
        category.append(temp_category)
```

```python
clcategory = []
for i in category:
    if (',' in i):
        clcategory.append(i[:-2])
    else:
        clcategory.append(i)
        
df_newattri = pd.DataFrame(name, columns=['name'])
df_newattri = pd.concat([df_newattri, pd.DataFrame(clcategory,columns=['category_name'])],axis=1)

print(df_newattri)
```

```python
# merge the new attribute to restaurantsDataFrame
RestaurantsDB = pd.merge(restaurantsDataFrame,df_newattri)
print(RestaurantsDB)
```

```python
# Find missing values
RestaurantsDB.isnull().any()
```

```python
# Check that attributes is unique or not
RestaurantsDB['id'].is_unique
```

```python
# delete duplicate recodes
RestaurantsDB = RestaurantsDB.drop_duplicates(subset = 'id',keep = 'first')
#reset index
RestaurantsDB = RestaurantsDB.reset_index(drop = True)
# check that attribute id is unique or nor again
RestaurantsDB['id'].is_unique
```
